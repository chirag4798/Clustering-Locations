```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tf = pd.read_csv('https://cocl.us/datascience_survey_data', index_col=0)
tf.sort_values('Very interested', axis=0, inplace=True, ascending=False)
```

```python
tf['Sum'] = tf['Very interested'] + tf['Somewhat interested'] + tf['Not interested']
tf
```

```python
columns = ['Very interested','Somewhat interested','Not interested']
```

```python
for i in columns:
    tf[i] = round(((tf[i]/tf['Sum']) * 100), 2)
```

```python
tf.drop('Sum', axis=1, inplace=True)
```

```python
tf
```

```python
ax = tf.plot(kind='bar', 
                figsize=(20, 8),
                color = ['#5cb85c','#5bc0de','#d9534f'],
                width=0.8,
                fontsize=14)

ax.set_title('Percentage of Respondents Interest in Data Science Areas',fontsize=16)

ax.set_facecolor('white')

ax.legend(fontsize=14,facecolor = 'white')

ax.get_yaxis().set_visible(False)

for p in ax.patches:
    ax.annotate(np.round(p.get_height(),decimals=2), 
                (p.get_x()+p.get_width()/2., p.get_height()), 
                ha='center', 
                va='center', 
                xytext=(0, 10), 
                textcoords='offset points',
                fontsize = 14
               )
    
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.legend(loc=(0.9,0.8))
```
